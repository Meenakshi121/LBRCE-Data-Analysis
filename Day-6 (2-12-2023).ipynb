### Line
* Line is a graph it shows frequency of data in a number line

```python
x=[10,12,14,16]
y=[4,5,6,7]
plt.plot(x,y,label="Line")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("Line graph")
plt.legend(loc="lower right")
```

```python
import numpy as np
x=np.linspace(0,5,20)
x
```

```python
plt.plot(x,x**2,color="red",label="Line1")
plt.plot(x,x**3,color="green",label="Line2")
plt.plot(x,x**4,color="yellow",label="Line3")
plt.legend()
```

```python
help(plt.plot)
```

```python
x1=np.arange(100)
x1
```

```python
plt.plot(x1,x1+10,color='c',linestyle='--')
plt.plot(x1,x1+20,color='m',linestyle='-.')
plt.plot(x1,x1+30,color='y',linestyle=':')
plt.plot(x1,x1+40,color='g',linestyle='-')
```

### Bar chart

* along with pyplot,we can use bar() to draw bar chart
* plt.bar()

```python
year=['mar 2020','april 2021','may 2022']
goldprices=[4500,5000,5500]
silver=[1000,1200,1500]
plt.bar(year,goldprices,width=0.33,color='green')
plt.bar(year,silver,width=0.33,color='red')
```

```python
x=[1,2,3,4,5,6]
y=[10,11,12,13,14,15]
plt.bar(x,y,color=['c','m','g','y','k','r'],edgecolor="blue")
```

```python
# horizontal barchart

# plt.barh()

plt.barh(x,y)
```

### boxpoint

* It gives the summary of numerical data through quartiles
* min
* Inter Quartile Range(IQR)
    * lower quartile(25%)
    * Median(50%)
    * Upper Quartile Range(75%)
* max()
* plt.boxplot()

```python
numbers=np.random.randint(0,1000,1000)
plt.boxplot(numbers)
```

```python
# Notched boxplot

plt.boxplot(numbers,notch=True)
```

### Histogram

* It is a diagram which shows the frequency distribution of numerical data
* plt.hist()

```python
x=[10,20,54,68,74,56,25]
plt.hist(x)
```

### Piechart

* plt.pie()


```python
games=["hockey","chess","tennis","football"]
scores=[92,85,79,70]
plt.pie(scores,labels=games,autopct="%1.2f%%",explode=[0,0.2,0,0],shadow=True)
plt.show()
plt.savefig("piechart.png")
```

### Categorical plot
* By default it returns scatter plot
* sns.catplot()

```python
sns.catplot(x="species",y="sepal_length",data=ir)
```

### Categorical distribution data
* boxplot

```python
sns.catplot(x="species",y="sepal_length",data=ir,kind="box")
```

```python
sns.catplot(data=ir,kind="box")
```

```python
sns.catplot(data=ir,kind="box",orient='h')
```

```python
sns.catplot(x="species",y="sepal_length",data=ir,kind="violin")
```

```python
sns.catplot(x="species",y="sepal_length",data=ir,kind="bar")
```

### Jointplot
* It is a combination of 2 plots, by default scatterplot , histogram
* sns.jointplot()

```python
sns.jointplot(x="species",y="sepal_length",data=ir,color='g')
```

```python
sns.jointplot(x="sepal_width",y="sepal_length",data=ir,color='g')
```
