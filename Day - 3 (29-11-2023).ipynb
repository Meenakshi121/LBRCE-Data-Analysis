```python
def grt(a,b):
    if a>b:
        return a
    else:
        return b
grt(87,909)
```

```python
vgrt=np.vectorize(grt)
vgrt([10,2,340],[90,89,87])
```

### Pandas

* Pandas stands for python data analysis library
* It is free and open source
* It is an excellent tool for Data Processing and analyzing real world data

### Pandas Data structures

* 1. Series : Used to create 1 dimensional array with named index eg:single column in an excel sheet
* 2. Data Frame: Used to create 2 Dimensional array with row index and column index eg:table in an excel sheet

```python
import pandas as pd
```

```python
pd.__version__
```

```python
# creating series using list
l=[10,20,30,40,50]
s=pd.Series(l)
print(s)
print(type(s))
```

```python
# changing the index
s.index=['a','b','c','d','e']
s
```

```python
s=pd.Series([90,'hello',70.8],index=['x','y','z'])
s
```

```python
# creating series by using numpy array

a=np.array([90,87,56])
s1=pd.Series(a)
s1
```

```python
# creating series by using tuple

s=pd.Series((100,294,45,65))
s
```

```python
# creating series by using dictionary

d={"stu1":90,"stu2":86,"stu3":88}
s=pd.Series(d)
s
```
